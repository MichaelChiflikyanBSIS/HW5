```python
import altair as alt
import pandas as pd

data_url = 'https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/bfro_reports_fall2022.csv'
df = pd.read_csv(data_url)
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df = df.dropna(subset=['date', 'longitude', 'latitude', 'state'])

alt.data_transformers.disable_max_rows()

plot1 = alt.Chart(data_url).mark_rect().encode(
    x=alt.X('classification:N', title='Report Classification'),
    y=alt.Y('state:N', title='State'),
    color=alt.Color('count():Q', scale=alt.Scale(scheme='viridis'), title='Number of Sightings'),
    tooltip=[
        alt.Tooltip('state:N'), 
        alt.Tooltip('classification:N'), 
        alt.Tooltip('count():Q')
    ]
).properties(
    width=400,
    height=600,
    title="Bigfoot Sightings Density by State"
)

brush = alt.selection_interval()

plot2 = alt.Chart(data_url).mark_circle().encode(
    x=alt.X('longitude:Q', scale=alt.Scale(domain=[-130, -60]), title='Longitude'),
    y=alt.Y('latitude:Q', scale=alt.Scale(domain=[20, 55]), title='Latitude'),
    color=alt.condition(brush, alt.Color('season:N', scale=alt.Scale(scheme='category10')), alt.value('lightgray')),
    tooltip=[
        alt.Tooltip('title:N'), 
        alt.Tooltip('date:T'), 
        alt.Tooltip('state:N')
    ]
).add_params(
    brush
).properties(
    width=600,
    height=400,
    title="Geographic Distribution of Sightings (Select area to highlight)"
)

plot1.display()
plot2.display()
```

```python
plot1.save('plot1.png')
plot2.save('plot2.png')
```
